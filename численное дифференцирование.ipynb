1. Машинное эпсилон


```python
from numpy import *
e = 1
i = 1
while 1 - e != 1:
    e = e/2
    i += 1
print("получили е =",e,"за",i,"итераций")
allclose(1,1+e)
```

3. Вычисление первой производной методом дифференцирования с комплексным шагом

```python
import cmath
def f1(x):
    return sin(x)
import matplotlib.pyplot as plt
xx = linspace(0, 1.5, 101)
plt.plot(xx, [f1(x) for x in xx])
```

```python
import cmath
h = 0.1j
f11 = [(f1(xx[m] + h).imag)/h.imag for m in range(len(xx))]
plt.plot(xx,f11,"bs")
plt.plot(xx,[f1(x) for x in xx])
plt.plot(xx,[cos(x) for x in xx],"^")
plt.show
```
